```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Datos extrados de la tabla
X = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26]).reshape(-1, 1)
y = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])

# Dividimos los datos en sets de entrenamiento y prueba (usaremos todos para entrenamiento por simplicidad)
X_train = X
y_train = y

# Creamos el objeto de regresin lineal
regr = linear_model.LinearRegression()

# Entrenamos el modelo usando los sets de entrenamiento
regr.fit(X_train, y_train)

# Hacemos predicciones usando los mismos datos de entrada (no tenemos un set de prueba separado en este caso)
y_pred = regr.predict(X_train)

# Imprimimos los coeficientes
print("Coeficientes: \n", regr.coef_)
# Error cuadrtico medio
print("Mean squared error: %.2f" % mean_squared_error(y_train, y_pred))
# Coeficiente de determinacin (R^2)
print("Coefficient of determination: %.2f" % r2_score(y_train, y_pred))

# Graficamos los puntos de datos y la lnea de regresin
plt.scatter(X_train, y_train, color="black")
plt.plot(X_train, y_pred, color="blue", linewidth=3)

plt.xlabel('xi')
plt.ylabel('yi')

plt.show()
```

El ejercicio que he realizado es un ejemplo de regresin lineal simple utilizando los datos proporcionados en la tabla de los restaurantes.

### 1. **Objetivo del ejercicio**
El objetivo es ajustar un modelo de regresin lineal que relacione dos variables:
- \( x_i \): la variable independiente, que podra ser el nmero de restaurantes u otra mtrica.
- \( y_i \): la variable dependiente, que podra representar los ingresos o ventas en esos restaurantes.

La regresin lineal trata de encontrar la relacin entre estas dos variables en la forma de una lnea recta que mejor se ajuste a los datos. La ecuacin de la lnea recta es:
\[
y = mx + b
\]
donde:
- \( y \) es el valor predicho de la variable dependiente (por ejemplo, ventas),
- \( m \) es la pendiente de la recta (coeficiente de regresin),
- \( x \) es el valor de la variable independiente (por ejemplo, nmero de restaurantes),
- \( b \) es la interseccin de la recta con el eje \( y \) (ordenada al origen).

## Frmulas:

### Pendiente \(b_1\):

$b_1 = \frac{\sum x_i y_i - \frac{\sum x_i \sum y_i}{n}}{\sum x_i^2 - \frac{(\sum x_i)^2}{n}}$

Esta frmula tambin puede expresarse como:

$b_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2} = \frac{S_{XY}}{S_{x}^2}$

Donde:

- $S_{XY}$ es la covarianza entre $x$ e $y$,
- $S_{x}^2$ es la varianza de $x$.

### Interseccin \(b_0\):

$b_0 = \bar{y} - b_1 \bar{x}$

Donde:

- $\bar{x}$ es la media de $x$,
- $\bar{y}$ es la media de $y$.

### Datos extrados de la tabla anterior:

- $y_i = [58, 105, 88, 118, 117, 137, 157, 169, 149, 202]$
- $\sum x_i = 140$
- $\sum y_i = 1300$
- $\sum x_i y_i = 21040$

Clculo de $b_1$ y $b_0$:


Primero, vamos a calcular los valores manualmente usando Python y luego dibujar la recta de regresin:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26])
y = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])

# Nmero de puntos
n = len(x)

# Sumas necesarias para las frmulas
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x_squared = np.sum(x ** 2)

# Medias de x e y
mean_x = np.mean(x)
mean_y = np.mean(y)

# Clculo de b1 (pendiente)
b1 = (sum_xy - (sum_x * sum_y) / n) / (sum_x_squared - (sum_x ** 2) / n)

# Clculo de b0 (interseccin con el eje y)
b0 = mean_y - b1 * mean_x

# Mostrar los coeficientes
print(f"Pendiente (b1): {b1}")
print(f"Interseccin (b0): {b0}")

# Valores predichos de y usando la ecuacin de la recta
y_pred = b0 + b1 * x

# Graficar los puntos y la recta de regresin
plt.scatter(x, y, color="black", label="Datos")
plt.plot(x, y_pred, color="blue", label="Recta de regresin")
plt.xlabel("xi")
plt.ylabel("yi")
plt.legend()
plt.show()
```

### Explicacin:
1. **Clculo de la pendiente (\(b_1\))**:
   - Usamos la frmula para \(b_1\), que relaciona la covarianza entre \(x\) e \(y\) con la varianza de \(x\). Este valor determina la inclinacin de la recta.
   
2. **Clculo de la interseccin (\(b_0\))**:
   - Despus de obtener \(b_1\), calculamos \(b_0\) usando la media de \(x\) e \(y\). Este valor representa dnde la recta corta el eje \(y\).

3. **Graficacin**:
   - Dibujamos los puntos proporcionados (\(x_i, y_i\)) como un diagrama de dispersin.
   - Luego graficamos la lnea de regresin ajustada, que representa la mejor estimacin lineal de \(y\) en funcin de \(x\).

Con esto, obtenemos la recta de regresin manualmente usando las frmulas tradicionales, lo cual nos da un entendimiento ms profundo del proceso detrs de la regresin lineal que las herramientas automticas como `scikit-learn` realizan internamente.

## Clculo de $r^2$ y $r$ usando los datos:

- $x_i = [2, 6, 8, 8, 12, 16, 20, 20, 22, 26]$
- $y_i = [58, 105, 88, 118, 117, 137, 157, 169, 149, 202]$

### 1. Promedios:

$$
\bar{x} = \frac{2 + 6 + 8 + 8 + 12 + 16 + 20 + 20 + 22 + 26}{10} = 14
$$

$$
\bar{y} = \frac{58 + 105 + 88 + 118 + 117 + 137 + 157 + 169 + 149 + 202}{10} = 130
$$

### 2. Clculo de $r^2$:

$$
r^2 = \frac{\left( \sum (x_i - \bar{x})(y_i - \bar{y}) \right)^2}{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}
$$

Aplicando los datos:

$$
r^2 = \frac{\left( 2(-72) + 6(-25) + 8(-42) + 8(-12) + 12(-13) + 16(7) + 20(27) + 20(39) + 22(19) + 26(72) \right)^2}{\left( (2 - 14)^2 + (6 - 14)^2 + ... + (26 - 14)^2 \right) \left( (58 - 130)^2 + (105 - 130)^2 + ... + (202 - 130)^2 \right)}
$$

Tras realizar los clculos:

$$
r^2 \approx 0.909
$$

### 3. Clculo de $r$:

$$
r = \sqrt{r^2} = \sqrt{0.909} \approx 0.953
$$

### Resultado final:

- $r^2 \approx 0.909$
- $r \approx 0.953$


## Clculo de los coeficientes de determinacin $r^2$ y $r$:

El coeficiente de determinacin, $r^2$, se calcula usando la siguiente frmula:

$$
r^2 = \frac{\left( \sum (x_i - \bar{x})(y_i - \bar{y}) \right)^2}{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}
$$

Donde:

- $x_i$ son los valores de la variable independiente,
- $y_i$ son los valores de la variable dependiente,
- $\bar{x}$ es la media de los valores $x_i$,
- $\bar{y}$ es la media de los valores $y_i$.

El coeficiente de correlacin, $r$, es la raz cuadrada del coeficiente de determinacin:

$$
r = \sqrt{r^2}
$$

### Interpretacin:

- Si $r^2 = 1$, significa que la relacin entre $x$ y $y$ es perfecta y lineal.
- Si $r^2 = 0$, significa que no hay relacin lineal entre $x$ y $y$.

### Frmulas adicionales:

La suma total de cuadrados ($SST$) se calcula como:

$$
SST = \sum (y_i - \bar{y})^2
$$

La suma de cuadrados explicada ($SSR$) es:

$$
SSR = \sum (\hat{y}_i - \bar{y})^2
$$

La suma de cuadrados del error ($SSE$) se calcula como:

$$
SSE = \sum (y_i - \hat{y}_i)^2
$$

Donde $\hat{y}_i$ son los valores predichos de $y$ usando la ecuacin de regresin.

El coeficiente de determinacin tambin puede expresarse como:

$$
r^2 = \frac{SSR}{SST}
$$

### Resumen:

- $r^2$ mide la proporcin de la varianza en $y$ que es explicada por la variable $x$.
- $r$ es el coeficiente de correlacin lineal entre $x$ y $y$.
