```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Datos extraídos de la tabla
X = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26]).reshape(-1, 1)
y = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])

# Dividimos los datos en sets de entrenamiento y prueba (usaremos todos para entrenamiento por simplicidad)
X_train = X
y_train = y

# Creamos el objeto de regresión lineal
regr = linear_model.LinearRegression()

# Entrenamos el modelo usando los sets de entrenamiento
regr.fit(X_train, y_train)

# Hacemos predicciones usando los mismos datos de entrada (no tenemos un set de prueba separado en este caso)
y_pred = regr.predict(X_train)

# Imprimimos los coeficientes
print("Coeficientes: \n", regr.coef_)
# Error cuadrático medio
print("Mean squared error: %.2f" % mean_squared_error(y_train, y_pred))
# Coeficiente de determinación (R^2)
print("Coefficient of determination: %.2f" % r2_score(y_train, y_pred))

# Graficamos los puntos de datos y la línea de regresión
plt.scatter(X_train, y_train, color="black")
plt.plot(X_train, y_pred, color="blue", linewidth=3)

plt.xlabel('xi')
plt.ylabel('yi')

plt.show()
```

El ejercicio que he realizado es un ejemplo de regresión lineal simple utilizando los datos proporcionados en la tabla de los restaurantes.

### 1. **Objetivo del ejercicio**
El objetivo es ajustar un modelo de regresión lineal que relacione dos variables:
- \( x_i \): la variable independiente, que podría ser el número de restaurantes u otra métrica.
- \( y_i \): la variable dependiente, que podría representar los ingresos o ventas en esos restaurantes.

La regresión lineal trata de encontrar la relación entre estas dos variables en la forma de una línea recta que mejor se ajuste a los datos. La ecuación de la línea recta es:
\[
y = mx + b
\]
donde:
- \( y \) es el valor predicho de la variable dependiente (por ejemplo, ventas),
- \( m \) es la pendiente de la recta (coeficiente de regresión),
- \( x \) es el valor de la variable independiente (por ejemplo, número de restaurantes),
- \( b \) es la intersección de la recta con el eje \( y \) (ordenada al origen).

### 2. **Explicación del código**
El código realiza los siguientes pasos:

#### a) **Importación de librerías**
Se importan las librerías necesarias para:
- Crear gráficos (`matplotlib`),

- Entrenar el modelo de regresión lineal (`linear_model` de `scikit-learn`),
- Evaluar el rendimiento del modelo (`mean_squared_error` y `r2_score`).

#### b) **Cargar los datos**
Se extraen los datos de la tabla que has proporcionado:
- La columna \( x_i \) tiene los valores: `[2, 6, 8, 8, 12, 16, 20, 20, 22, 26]`.
- La columna \( y_i \) tiene los valores: `[58, 105, 88, 118, 117, 137, 157, 169, 149, 202]`.

Estos datos representan el conjunto de observaciones que usaremos para ajustar nuestro modelo de regresión lineal.

#### c) **Entrenar el modelo**
Usamos la función `LinearRegression()` de `scikit-learn` para crear un objeto de regresión lineal y entrenarlo con los datos. El modelo ajusta la mejor línea que minimiza la suma de los errores cuadrados entre los valores reales de \( y_i \) y los valores predichos.

#### d) **Evaluación del modelo**
Después de entrenar el modelo, evaluamos su rendimiento:
- **Coeficiente de regresión**: Representa la pendiente de la línea, es decir, cuánto cambia \( y \) por cada incremento unitario en \( x \).
- **Error cuadrático medio (MSE)**: Es una métrica que mide el promedio de los errores al cuadrado entre las predicciones del modelo y los valores reales. Un valor menor significa que el modelo predice con mayor precisión.
- **Coeficiente de determinación (R²)**: Mide qué tan bien se ajustan los datos al modelo. Un valor de \( R^2 \) de 1 indica un ajuste perfecto.

#### e) **Visualización del modelo**
Finalmente, se crea un gráfico que muestra:
- Los puntos negros representan los datos originales (\( x_i, y_i \)),
- La línea azul muestra la línea de regresión ajustada por el modelo.

## Fórmulas:

### Pendiente \(b_1\):

$b_1 = \frac{\sum x_i y_i - \frac{\sum x_i \sum y_i}{n}}{\sum x_i^2 - \frac{(\sum x_i)^2}{n}}$

Esta fórmula también puede expresarse como:

$b_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2} = \frac{S_{XY}}{S_{x}^2}$

Donde:

- $S_{XY}$ es la covarianza entre $x$ e $y$,
- $S_{x}^2$ es la varianza de $x$.

### Intersección \(b_0\):

$b_0 = \bar{y} - b_1 \bar{x}$

Donde:

- $\bar{x}$ es la media de $x$,
- $\bar{y}$ es la media de $y$.

### Datos extraídos de la tabla anterior:
- $x_i = [2, 6, 8, 8, 12, 16, 20, 20, 22, 26]$
- $y_i = [58, 105, 88, 118, 117, 137, 157, 169, 149, 202]$
- $\sum x_i = 140$
- $\sum y_i = 1300$
- $\sum x_i y_i = 21040$
- $\sum x_i^2 = 2528$


Cálculo de $b_1$ y $b_0$:


Primero, vamos a calcular los valores manualmente usando Python y luego dibujar la recta de regresión:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26])
y = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])

# Número de puntos
n = len(x)

# Sumas necesarias para las fórmulas
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x_squared = np.sum(x ** 2)

# Medias de x e y
mean_x = np.mean(x)
mean_y = np.mean(y)

# Cálculo de b1 (pendiente)
b1 = (sum_xy - (sum_x * sum_y) / n) / (sum_x_squared - (sum_x ** 2) / n)

# Cálculo de b0 (intersección con el eje y)
b0 = mean_y - b1 * mean_x

# Mostrar los coeficientes
print(f"Pendiente (b1): {b1}")
print(f"Intersección (b0): {b0}")

# Valores predichos de y usando la ecuación de la recta
y_pred = b0 + b1 * x

# Graficar los puntos y la recta de regresión
plt.scatter(x, y, color="black", label="Datos")
plt.plot(x, y_pred, color="blue", label="Recta de regresión")
plt.xlabel("xi")
plt.ylabel("yi")
plt.legend()
plt.show()
```

### Explicación:
1. **Cálculo de la pendiente (\(b_1\))**:
   - Usamos la fórmula para \(b_1\), que relaciona la covarianza entre \(x\) e \(y\) con la varianza de \(x\). Este valor determina la inclinación de la recta.
   
2. **Cálculo de la intersección (\(b_0\))**:
   - Después de obtener \(b_1\), calculamos \(b_0\) usando la media de \(x\) e \(y\). Este valor representa dónde la recta corta el eje \(y\).

3. **Graficación**:
   - Dibujamos los puntos proporcionados (\(x_i, y_i\)) como un diagrama de dispersión.
   - Luego graficamos la línea de regresión ajustada, que representa la mejor estimación lineal de \(y\) en función de \(x\).

Con esto, obtenemos la recta de regresión manualmente usando las fórmulas tradicionales, lo cual nos da un entendimiento más profundo del proceso detrás de la regresión lineal que las herramientas automáticas como `scikit-learn` realizan internamente.

## Cálculo de $r^2$ y $r$ usando los datos:

- $x_i = [2, 6, 8, 8, 12, 16, 20, 20, 22, 26]$
- $y_i = [58, 105, 88, 118, 117, 137, 157, 169, 149, 202]$

### 1. Promedios:

$$
\bar{x} = \frac{2 + 6 + 8 + 8 + 12 + 16 + 20 + 20 + 22 + 26}{10} = 14
$$

$$
\bar{y} = \frac{58 + 105 + 88 + 118 + 117 + 137 + 157 + 169 + 149 + 202}{10} = 130
$$

### 2. Cálculo de $r^2$:

$$
r^2 = \frac{\left( \sum (x_i - \bar{x})(y_i - \bar{y}) \right)^2}{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}
$$

Aplicando los datos:

$$
r^2 = \frac{\left( 2(-72) + 6(-25) + 8(-42) + 8(-12) + 12(-13) + 16(7) + 20(27) + 20(39) + 22(19) + 26(72) \right)^2}{\left( (2 - 14)^2 + (6 - 14)^2 + ... + (26 - 14)^2 \right) \left( (58 - 130)^2 + (105 - 130)^2 + ... + (202 - 130)^2 \right)}
$$

Tras realizar los cálculos:

$$
r^2 \approx 0.909
$$

### 3. Cálculo de $r$:

$$
r = \sqrt{r^2} = \sqrt{0.909} \approx 0.953
$$

### Resultado final:

- $r^2 \approx 0.909$
- $r \approx 0.953$


## Cálculo de los coeficientes de determinación $r^2$ y $r$:

El coeficiente de determinación, $r^2$, se calcula usando la siguiente fórmula:

$$
r^2 = \frac{\left( \sum (x_i - \bar{x})(y_i - \bar{y}) \right)^2}{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}
$$

Donde:

- $x_i$ son los valores de la variable independiente,
- $y_i$ son los valores de la variable dependiente,
- $\bar{x}$ es la media de los valores $x_i$,
- $\bar{y}$ es la media de los valores $y_i$.

El coeficiente de correlación, $r$, es la raíz cuadrada del coeficiente de determinación:

$$
r = \sqrt{r^2}
$$

### Interpretación:

- Si $r^2 = 1$, significa que la relación entre $x$ y $y$ es perfecta y lineal.
- Si $r^2 = 0$, significa que no hay relación lineal entre $x$ y $y$.

### Fórmulas adicionales:

La suma total de cuadrados ($SST$) se calcula como:

$$
SST = \sum (y_i - \bar{y})^2
$$

La suma de cuadrados explicada ($SSR$) es:

$$
SSR = \sum (\hat{y}_i - \bar{y})^2
$$

La suma de cuadrados del error ($SSE$) se calcula como:

$$
SSE = \sum (y_i - \hat{y}_i)^2
$$

Donde $\hat{y}_i$ son los valores predichos de $y$ usando la ecuación de regresión.

El coeficiente de determinación también puede expresarse como:

$$
r^2 = \frac{SSR}{SST}
$$

### Resumen:

- $r^2$ mide la proporción de la varianza en $y$ que es explicada por la variable $x$.
- $r$ es el coeficiente de correlación lineal entre $x$ y $y$.
